```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_csv("files/titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop("cabin",axis=1, inplace=True)
df.head()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
df ["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])
df.isnull().sum()
```

Feature Engineering

```python
df["family_size"] = df["sibSp"] + df["parch"] + 1
```

```python
df["title"] = df["name"].str.extract(" ([A-Za-z]+)\.",expand= False)
df["title"].value_counts()
```

```python
df["title"] = df["title"].replace(["Dr" ,"Rev","Mlle","Major","Col","Countess","Capt","Ms","Sir","Lady","Mme","Don","Jonkheer"],"Rare")

df["title"] = df["title"].replace({"Mlle": "Miss" , "Ms":"Miss" , "Lady" : "Mrs" ,"Mme" : "Mrs"})
df["title"].value_counts()
```

```python
df =df.drop(["passengerId","name","ticket"],axis=1)
df.head()
```

```python
df.describe()
```

```python

label_cols = ['sex', 'embarked', 'title']
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

```python
num_cols = ['age', 'fare', 'family_size']
scaler = RobustScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
df.head()
```

```python
df.describe()
```

```python
before = len(df)
df = df.drop_duplicates()
after = len(df)

removed = before - after
removed
```

```python
df.to_csv("files/cleaned_titanic.csv",index=False)
```
